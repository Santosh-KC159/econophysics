```python
def preprocess_data_for_labels(ticker):
    days = 7
    df = pd.read_csv('data/sp500_joined_closes.csv')
    tickers = df.columns.values.tolist()
    df.fillna(0, inplace = True)
    
    for i in range(1, days+1):
        df['{}_{}'.format(ticker, i)] = (df[ticker].shift(-1) - df[ticker]) / df[ticker]
        
    df.fillna(0, inplace=True)
    return tickers, df
```

```python
def sell_buy_hold(*args):
    cols = [c for c in args]
    requirement = 0.02
    for col in cols:
        if (col > requirement):
            return 1
        if (col < -requirement):
            return -1
    return 0
```
